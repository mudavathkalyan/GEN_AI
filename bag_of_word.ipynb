```python
import pandas as pd
```

```python
df=pd.read_csv('spam.csv')
```

```python
df.Category.value_counts()
```

```python
# creating new column
df['spam']=df['Category'].apply(lambda x:1 if x=='spam' else 0)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(df.Message,df.spam,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()

X_train_cv=v.fit_transform(X_train)

```

```python
X_train_cv.toarray() # vectors
```

```python
#Building  model
```

```python
from sklearn.naive_bayes import MultinomialNB

model=MultinomialNB()
model.fit(X_train_cv,y_train)
```

```python
X_test_cv=v.transform(X_test)
```

```python
from sklearn.metrics import classification_report

y_pred=model.predict(X_test_cv)

print(classification_report(y_test,y_pred))
```

```python
#now check
emails=[
    'hi kalyan how r u?',
    'hey kalyan 20% free,and 10% discount',
    'Do visit out store'
]
```

```python
emails_count=v.transform(emails)
model.predict(emails_count)
```

```python
#observe the above message are classified into spam and not spam 
#based on the initially provided data
#so 
```

```python
# Easy way another using pipline
from sklearn.pipeline import Pipeline

clf=Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
classification_report(y_test,y_pred)
```

```python
#now check
emails=[
    'hi kalyan how r u?',
    'hey kalyan 20% free,and 10% discount',
    'Do visit out store'
]
```

```python
predictions = clf.predict(emails)
```

```python
predictions
```

```python
# Print results
for email, pred in zip(emails, predictions):
    label = "Spam" if pred == 1 else "Not Spam"
    print(f"Email: '{email}' → Prediction: {label}")
```
