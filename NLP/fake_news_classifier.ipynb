```python
from tensorflow.keras.layers import Embedding # helps like word 2 vec
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
#one hot 
messages=X.copy()
```

```python
messages['title'][1]
```

```python
messages
```

```python
import nltk
import re #regular exp
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer ##stemming purpose
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages.iloc[i]['title'])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
voc_size=5000
```

```python
onehot_rep=[one_hot(words,voc_size) for words in corpus]
onehot_rep
```

```python
corpus[1]
```

```python
onehot_rep[1] #index of those words
```

## Embedding Representation

```python
sent_length=20
embedded_docs=pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
```

```python
print(embedded_docs)
```

```python
embedded_docs[1]
```

## creating model

```python
embedding_vector_features=40 #every feature representtion into 40 dimens
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
embedded_docs.shape,y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_final,y_final,test_size=0.3,random_state=42)
```

```python
## finally training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

## performance metrices  and Accuracy

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred>0.5,1,0)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
import pickle
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import load_model

# Load Tokenizer
with open('tokenizer.pkl', 'rb') as handle:
    tokenizer = pickle.load(handle)

# Load Trained Model
model = load_model('fake_news_model.h5')  # Ensure the model is saved as 'fake_news_model.h5'

# Define max length (must be same as during training)
MAX_LENGTH = 100  # Change if different in training

def preprocess_text(text):
    """Function to preprocess input text similar to training data."""
    import re
    from nltk.corpus import stopwords
    from nltk.stem.porter import PorterStemmer

    ps = PorterStemmer()
    text = re.sub('[^a-zA-Z]', ' ', text)  # Remove special characters
    text = text.lower().split()
    text = [ps.stem(word) for word in text if word not in stopwords.words('english')]
    return ' '.join(text)

def predict_news(review):
    """Function to predict whether news is fake or real."""
    processed_review = preprocess_text(review)  # Preprocess the text
    sequence = tokenizer.texts_to_sequences([processed_review])  # Convert to sequence
    padded_sequence = pad_sequences(sequence, maxlen=MAX_LENGTH, padding='post')  # Pad sequence
    
    prediction = model.predict(padded_sequence)[0][0]  # Get model prediction
    return "Fake News" if prediction > 0.5 else "Real News"

```

```python
# Example Usage
news_title = input("Enter a news headline: ")
result = predict_news(news_title)
print(f"Prediction: {result}")
```
