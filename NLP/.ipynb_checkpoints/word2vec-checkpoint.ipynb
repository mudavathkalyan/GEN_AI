```python
import gensim
from gensim.models import Word2Vec

# Sample corpus (list of tokenized sentences)
corpus = [
    ["hello", "world","happy","enjoy","smile"],
    ["machine", "learning", "is", "awesome","sad"],
    ["natural", "language", "processing", "is", "cool"],
    ["hello", "machine", "world"]
]

# Create and train a Word2Vec model
model = Word2Vec(sentences=corpus, vector_size=100, window=3, min_count=1, workers=4)

# Save the model
model.save("word2vec_example.model")

# Load the model
model = Word2Vec.load("word2vec_example.model")

# Get vector representation of a word
vector = model.wv["hello"]
print(f"Vector for 'hello': {vector}")

# Find most similar words to a given word
similar_words = model.wv.most_similar("machine", topn=3)
print(f"Words similar to 'machine': {similar_words}")
```
